#### Bais and Variance
- *Bais*: 
1. Bais is error introdice in ML model. 
2. High Bais means big difference between actual and predicted values. This is not good for model. 
3. Low Bais indicate difference between actual and predicted values is low. 
- *Variance*:
1. Variance indicate how Scattered data is. 
2. High variance is more Scattered data and Low variance indicates less scatttered data. 
#### Overfitting and Underfitting
- *Overfitting*:
1. overfitting occur when low bais and high variance in the data.
2. Overfitting happen when a model focus on to many details in the training data. 
3. *Has a negative impact on the performance of the model on new dataset.*
- *Undefitting*:
1. Underfitting occur when high bais and high variance in the data. 
2. It is detectable as it exhibit poor performance on the training data.
