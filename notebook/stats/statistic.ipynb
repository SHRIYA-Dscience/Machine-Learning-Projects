# Statsmodels

```python
import pandas as pd 
import numpy as np 
import statsmodels.api as sm 
import statsmodels.formula.api as smf
```

```python
data = sm.datasets.get_rdataset("Guerry", "HistData").data
```

```python
data.head()
```

```python
# Fit OLS Regression 
res = smf.ols('Lottery ~ Literacy + np.log(Pop1831)', data=data).fit()
```

```python
print(res.summary())
```

```python
# Using numpy array 
nobs = 100 
X = np.random.random((nobs,2))
X = sm.add_constant(X)
```

```python
beta = [1,.1,.5] # coef
e = np.random.random(nobs)
y = np.dot(X,beta) + e
```

```python
res_2 = sm.OLS(y,X).fit()
```

```python
print(res_2.summary())
```
